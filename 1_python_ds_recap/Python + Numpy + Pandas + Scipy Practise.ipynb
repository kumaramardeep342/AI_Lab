### Calculator

```python
%%writefile calculator.py
class Calculator:
    def __init__(self, a, b)->None:
        self.a = a
        self.b = b

    def add(self)->float:
        return self.a+self.b

    def substract(self)->float:
        return self.a-self.b

    def multiply(self)->float:
        return self.a*self.b

    def division(self)->float:
        return (self.a/self.b)

    def operation(self,op)->None:
        if (op == "add"):
            print(f"The sum of {self.a} and {self.b} is {self.add()}")
        elif (op == "substract"):
            print(f"The difference of {self.a} and {self.b} is {self.substract()}")
        elif (op == "multiply"):
            print(f"The product of {self.a} and {self.b} is {self.multiply()}")
        elif (op == "division"):
            if self.b == 0:
                raise ZeroDivisionError("Division by zero is not allowed")
            print(f"The division of {self.a} and {self.b} is {self.division()}")
        else :
            raise Exception("Please enter the operation correctly")
```

```python

from calculator import Calculator

if __name__ == "__main__":
    print("Welcome to Calculator...")
    try :

        a = input("Enter the first number: ")
        if not a.isnumeric():
            raise Exception("Please enter the first number correctly")
        logger.info(f"The first number is {a}")

        b = input("Enter the second number:")
        if not b.isnumeric():
            raise Exception("Please enter the second number correctly")
        logger.info(f"The second number is {b}")

        cal = Calculator(float(a), float(b))

        op = input("Enter the operation to be performed:  add, substract, multiply, division:, try to give the exact keyword: ")
        cal.operation(op)
        logger.info(f"The operation {op} is performed successfully...")

    except Exception as e :
        print(f"{e}")
        logger.error(f"The operation  is not performed successfully..")
```

```python
%%writefile test_calculator.py
import pytest
import calculator

def test_case_1():
    a = 12
    b = 13
    cal = calculator.Calculator(a,b)
    assert cal.add() == 25

def test_case_2():
    a = 12
    b = 0
    cal = calculator.Calculator(a,b)
    with pytest.raises(ZeroDivisionError):
        cal.division()

@pytest.mark.xfail
def test_case_3():
    a = 12
    b =  0
    cal = calculator.Calculator(a,b)
    assert cal.division() == 25

@pytest.fixture
def input_4():
    a = 12.0
    b = 13.0
    return a,b

def test_case_4(input_4):
    a,b = input_4
    cal = calculator.Calculator(a,b)
    assert cal.substract() == -1
```
