```python
#Method
class Car:
    def __init__(self, model, number):
        self.model = model
        self.number = number
    def mod_num(self):
        print(f"The car model is {self.model} and number is {self.number}")
```

```python
#Method examples
car = Car("toyota",1234)
car.model
```

```python
car.mod_num()
```

```python
#Inheritance
class Car:
    def __init__(self, model, number):
        self.model = model
        self.number = number
    def mod_num(self,model):
        return model

class Engine(Car):
    def __init__(self, model, engine):
        self.engine = engine
        self.model = model
    def mod_engine(self):
        print(f"The car model is {self.mod_num(self.model)} and number is {self.engine}")
```

```python
Engine('toyota','v6').mod_engine()
```

```python
#Encapsulation
class Car:
    def __init__(self):
        self.__model = 'toyota'
    def sell(self):
        print(f"I am selling my {self.__model} car")
        
car = Car()
car.__model = 1000
car.sell()
```

```python
#Polymorphism
class Car:
    def __init__(self):
        self.model = 'toyota'
    def sell(self):
        print(f"I am selling my {self.model} car")
class Maruthi(Car):
    def __init__(self):
        self.model = 'maruthi'
    def sell(self):
        print(f"I am selling my {self.model} car")
        
Maruthi().sell()
```

# Speed of numpy vs python

```python
# Data setup
N = 1_000_000
a = np.random.randn(N)   # NumPy array
la = a.tolist()          # Python list version (same values)

# Functions to benchmark
def py_filter():
    return [x for x in la if x > 0]

def np_filter():
    return a[a > 0]

print("1) Python list (list comprehension over la):")
%timeit py_filter()

print("\n2) NumPy array (boolean mask over a):")
%timeit np_filter()
```

```python
import numpy as np

# Synthetic regression data
n_samples = 200_000
n_features = 50

X = np.random.randn(n_samples, n_features)
y = np.random.randn(n_samples)

w = np.random.randn(n_features)
b = 0.1
lr = 1e-3

# Python-list versions of the same data
XL = X.tolist()
yL = y.tolist()
wL = w.tolist()


def predict_python():
    # Returns list of predictions
    out = []
    for i in range(n_samples):
        s = b
        row = XL[i]
        for j in range(n_features):
            s += row[j] * wL[j]
        out.append(s)
    return out

def predict_numpy():
    return X @ w + b


print("1) Python loops: linear regression prediction (list of lists):")
%timeit predict_python()

print("\n2) NumPy vectorized: linear regression prediction (X @ w + b):")
%timeit predict_numpy()
```

# Scipy

```python
from scipy import constants, special, integrate, optimize, interpolate, fftpack, signal, linalg, stats
import numpy as np

# 1. Constants & Special Functions
print(f"Pi: {constants.pi}, Speed of Light: {constants.c}")
print(f"Bessel Function (j0): {special.jn(0, 1.0)}")

# 2. Integration (integrate)
val, err = integrate.quad(lambda x: x**2, 0, 1)
print(f"Integral of x^2 [0,1]: {val:.4f}")

# 3. Linear Algebra (linalg)
A = np.array([[1, 2], [3, 4]])
print("Determinant:", linalg.det(A))
print("Inverse:\n", linalg.inv(A))

# 4. Fourier Transforms (fftpack)
sig = np.array([1, 2, 1, -1, 1.5])
print("FFT:", fftpack.fft(sig))

# 5. Signal Processing (signal)
print("Resampled Signal:", signal.resample(sig, 10))
```

### Interpolation & Statistics
- `interpolate`: Smoothing data.
- `stats`: Statistical distributions.

```python
# Interpolation
x = np.arange(10)
y = np.sin(x)
f = interpolate.interp1d(x, y, kind='cubic')
print("Interpolated value at 4.5:", f(4.5))

# Statistics
print("Normal PDF at 0:", stats.norm.pdf(0))
print("Random Normal Samples:", stats.norm.rvs(size=5))
```

```python
from scipy.optimize import minimize, curve_fit
import matplotlib.pyplot as plt

# --- 1. Optimization Example ---
# Objective: Minimize f(x) = x^2 + 10*sin(x)
def f(x):
    return x**2 + 10*np.sin(x)

res = minimize(f, x0=0)
print(f"Global Minimum found at x = {res.x[0]:.4f}")

# --- 2. Regression Example (Curve Fitting) ---
# Data: y = a * exp(b * x) + noise
def model(x, a, b):
    return a * np.exp(b * x)

x_data = np.linspace(0, 2, 50)
y_data = model(x_data, 2.5, 1.3) + 0.5 * np.random.normal(size=50)

popt, _ = curve_fit(model, x_data, y_data)
print(f"Regression Parameters: a={popt[0]:.2f}, b={popt[1]:.2f}")

# Visualization
plt.scatter(x_data, y_data, label='Data')
plt.plot(x_data, model(x_data, *popt), 'r', label='Fitted Curve')
plt.legend()
plt.title("Non-linear Regression with SciPy")
plt.show()
```
