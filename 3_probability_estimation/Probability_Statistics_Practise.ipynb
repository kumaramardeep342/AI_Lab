## Part 10: Parameter Estimation
### 10.1 Introduction to Parameter Estimation

```python
print("="*70)
print('Example : Estimating population mean and variance from sample')
print("="*70)

# setting seed for reproducibilty
np.random.seed(42)

# True population parameters (unknown in real scenarios)
mu, sigma = 50, 10 # mean and standard deviation

#drawing the sample
sample_size = 100
sample = np.random.normal(loc=mu, scale=sigma, size=sample_size)
print(f"10 sample is :\n {sample[:10]}")
print()

#point estimators
mu_hat = np.mean(sample)
var_hat = np.var(sample,ddof=1)
sigma_hat = np.std(sample,ddof = 1)

print('Parameter Estimation Example:')
print('-'*70)
print(f"Sample size = {sample_size}")
print(f"\nTrue Parameters Population:")

print(f"μ(mean) = {mu}")
print(f"σ² (variance) = {sigma**2}")
print(f"σ (std dev) = {sigma}")

print(f"\nEstimated Parameters (from sample):")
print(f"μ̂(mean) = {mu_hat:.2f}")
print(f"σ̂² (variance) = {var_hat:.2f}")
print(f"σ̂ (std dev) = {sigma_hat:.2f}")

print(f"\nEstimation Error:")
print(f"Mean Error : {abs(mu - mu_hat):.2f}")
print(f"Variance Error : {abs(sigma**2 - var_hat):.2f}")
```
